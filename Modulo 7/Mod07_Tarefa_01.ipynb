````python
metadata = pd.DataFrame({
    'variavel': df.columns,
    'tipo_dado': df.dtypes
})
````

````python
qtd_categorias = []
````

````python
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'bool':
        qtd_categorias.append(df[col].nunique())
    else:
        qtd_categorias.append(None)
````

````python
metadata['qtd_categorias'] = qtd_categorias
````

````python
df_encoded = pd.get_dummies(df, drop_first=True)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
categorical_vars = ['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']
````

````python
df_encoded = pd.get_dummies(df, columns=categorical_vars, drop_first=True)
````

````python
final_vars = [col for col in df_encoded.columns if df_encoded[col].dtype in ['int64', 'float64', 'uint8', 'bool']]
````

````python
df_final = df_encoded[final_vars]
````

````python
df_final.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
crosstab_email = pd.crosstab(df['possui_email'], df['mau'])
crosstab_veiculo = pd.crosstab(df['posse_de_veiculo'], df['mau'])
````

````python
proportion_email = crosstab_email.div(crosstab_email.sum(axis=1), axis=0)
proportion_veiculo = crosstab_veiculo.div(crosstab_veiculo.sum(axis=1), axis=0)
````

````python
print("Proporção de `mau = True` por `possui_email`:")
if True in proportion_email.columns:
    print(proportion_email[True])
else:
    print("Nenhum valor verdadeiro para 'mau' encontrado para `possui_email`.")
````

````python
print("\nProporção de `mau = True` por `posse_de_veiculo`:")
if True in proportion_veiculo.columns:
    print(proportion_veiculo[True])
else:
    print("Nenhum valor verdadeiro para 'mau' encontrado para `posse_de_veiculo`.")
````
