```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

```python
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
colunas_interesse = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']
sinasc_graficos = sinasc[colunas_interesse]
```

```python
for coluna in sinasc_graficos.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=coluna, data=sinasc_graficos, palette='viridis')
    plt.title(f'Distribuição de {coluna}')
    plt.xlabel(coluna)
    plt.ylabel('Contagem')
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
apg5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6)) 
sns.barplot(x='CONSULTAS', y='APGAR5', data=apg5_por_consultas, palette='muted') 
plt.title('Média de APGAR5 por Consultas')
plt.xlabel('Consultas') 
plt.ylabel('Média APGAR5') 
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
colunas_interesse = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for coluna, ax in zip(colunas_interesse, axs.flatten()):
    # Conta a frequência de cada valor na coluna
    frequencia = sinasc[coluna].value_counts()
    # Plota o gráfico de barras
    frequencia.plot(kind='bar', ax=ax, color='skyblue')
    ax.set_title(f'Frequência de {coluna}')
    ax.set_xlabel(coluna)
    ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```
