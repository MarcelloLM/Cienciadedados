```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))

sns.boxplot(x=sinasc['PESO'], color='skyblue')

plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso (g)')
plt.ylabel('')

plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))

plt.hist(sinasc['IDADEMAE'], bins=30, color='skyblue', edgecolor='black')

plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade')
plt.ylabel('Frequência')

plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6))

plt.hist(sinasc['PESO'], bins=30, color='skyblue', edgecolor='black')

plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso (g)')
plt.ylabel('Frequência')

plt.show()
```
